## The Game: Tic_Tac_Toe

```python
Layout = {'7': ' ' , '8': ' ' , '9': ' ' ,
          '4': ' ' , '5': ' ' , '6': ' ' ,
          '1': ' ' , '2': ' ' , '3': ' ' }

numbers = []

for key in Layout:
    numbers.append(key)
    

def Draw(lines):
    
    print(lines['7'] + '|' + lines['8'] + '|' + lines['9'])
    
    print('-+-+-')
    
    print(lines['4'] + '|' + lines['5'] + '|' + lines['6'])
    
    print('-+-+-')
    
    print(lines['1'] + '|' + lines['2'] + '|' + lines['3'])


                             # Gameplay
def game():

    turn = 'X'
    c    = 0
    
    for i in range(10):
        Draw(Layout)
        print("Your turn!," + turn +" "+ "Choose your place.")
        
        move = input()        
        
        if Layout[move] == ' ':
            Layout[move] = turn
            c  += 1
        else:
            print("That's Taken!\nChoose a New place:)")
            continue
        
        if c >= 5:
            if Layout['7'] == Layout['8'] == Layout['9'] != ' ': # across the top
                Draw(Layout)
                print("\nGame Over!!\n")
                print(" **** " +turn + " Won. ****")      
                break
                
            elif Layout['4'] == Layout['5'] == Layout['6'] != ' ': # across the middle
                Draw(Layout)
                print("\nGame Over!!\n")                
                print(" **** " +turn + " Won. ****")        
                break
            
            elif Layout['1'] == Layout['2'] == Layout['3'] != ' ': # across the bottom
                Draw(Layout)
                print("\nGame Over!!\n")                
                print(" **** " +turn + " Won. ****")       
                break
            
            elif Layout['1'] == Layout['4'] == Layout['7'] != ' ': # down the left side
                Draw(Layout)
                print("\nGame Over!!\n")                
                print(" **** " +turn + " Won. ****")        
                break
                
            elif Layout['2'] == Layout['5'] == Layout['8'] != ' ': # down the middle
                Draw(Layout)
                print("\nGame Over!!\n")                
                print(" **** " +turn + " Won. ****")
                break
            
            elif Layout['3'] == Layout['6'] == Layout['9'] != ' ': # down the right side
                Draw(Layout)
                print("\nGame Over!!\n")               
                print(" **** " +turn + " Won. ****")
                break 
            
            elif Layout['7'] == Layout['5'] == Layout['3'] != ' ': # diagonal
                Draw(Layout)
                print("\nGame Over!!\n")                
                print(" **** " +turn + " Won. ****")
                break
            
            elif Layout['1'] == Layout['5'] == Layout['9'] != ' ': # diagonal
                Draw(Layout)
                print("\nGame Over!!\n")                
                print(" **** " +turn + " Won. ****")
                break 
        
        # Change the player after every move.
        if turn =='X':
            turn = 'O'
        else:
            turn = 'X'
            
        # the board is full, declare  'tie'.
        if c == 9:
            print("\nGame Over!!\n")                
            print("It's a Tie!!")
        
        
    
    # Restart the Game
    restart = input("Do want to play Again?(y/n)")
    if restart == "y" or restart == "Y":  
        for key in numbers:
            Layout[key] = " "
            game()    
            
if __name__ =="__main__":
        game()
```
