### The Code

```python
from pytube import YouTube

# input statement
link = input("Enter your link:  ")

# Creating an object
YT   = YouTube(link)

# making sure of the best possible resolution
ys   =YT.streams.get_highest_resolution()

# While the task is under process
print("Downloading......")

# storage point of the downloaded video
ys.download("C:/Users/Byragi/Videos/Captures")


#Done
print("Download...... completed!")
```
